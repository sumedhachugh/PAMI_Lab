```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.externals.six import StringIO
import pydotplus
```

```python
iris_data = pd.read_csv("/Users/sumedha/Desktop/iris.csv")
iris_data.head()
```

```python
test_data1 = iris_data.iloc[0:10]
test_data2 = iris_data.iloc[50:60]
test_data3 = iris_data.iloc[100:110]
test_data = pd.concat([test_data1,test_data2,test_data3])
test_data
```

```python
train_data1 = iris_data.iloc[10:50]
train_data2 = iris_data.iloc[60:100]
train_data3 = iris_data.iloc[110:150]
train_data = pd.concat([train_data1,train_data2,train_data3])
train_data
```

```python
train_data_lastColumn = train_data.iloc[:,4]
train_data_lastColumn.head()
```

```python
test_data_lastColumn = test_data.iloc[:,4]
test_data_lastColumn.head()
```

```python
train_data_remainingColumn = train_data.iloc[:,:4]
train_data_remainingColumn.head()
```

```python
test_data_remainingColumn = test_data.iloc[:,:4]
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(train_data_remainingColumn, train_data_lastColumn)
```

```python
clf.predict([[4.3,3.0,1.1,0.1]])
```

```python
clf.predict(train_data_remainingColumn)
```

```python
tree.plot_tree(clf);
```

```python
fn=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
cn=['1', '2', '3']
dot_data = StringIO()
tree.export_graphviz(clf,
                     out_file = dot_data,
                     feature_names=fn,
                     class_names= cn,
                     filled=True, rounded=True,impurity = False)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
```

```python
graph.write_pdf("iris_data.pdf")
```
